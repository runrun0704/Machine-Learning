```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv('Bank.csv')
df
```

```python
str_list = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month']
str_df = df[str_list]
str_df
```

```python
str_df2 = pd.get_dummies(str_df, drop_first = True) 
num_df = df.drop(str_list, axis = 1)
df2 = pd.concat([num_df, str_df2, str_df], axis = 1)
df2
```

```python
train_val, test = train_test_split(df2, test_size = 0.1, random_state = 0)
train_val
```

```python
from sklearn import tree

def learn(x, t, j):
    x_train, x_test, y_train, y_test = train_test_split(x, t,test_size = 0.2,random_state = 0)
    model = tree.DecisionTreeClassifier(max_depth = j, random_state = 0, class_weight = 'balanced')
    model.fit(x_train, y_train)
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    return train_score, test_score, model
```

```python
def score(a):
    for i in range(1, 20):
        s1, s2 , model = learn(a, t, i)
        print(i, s1, s2)
```
