```python
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_csv('Bank.csv')
df.head(3)
```

```python
str_cor = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact','month']
str_df = df[str_cor]
str_df
```

```python
str_df2 = pd.get_dummies(str_df, drop_first = True)
num_df = df.drop(str_cor, axis  = 1)
df2 = pd.concat([num_df, str_df2, str_df], axis = 1)
df2
```

```python
from sklearn import tree
```

```python
def learn_tree(x, t, i):
    x_train, x_val, y_train, y_val = train_test_split(x, t,test_size = 0.2,random_state = 0)
    model_tree = tree.DecisionTreeClassifier(max_depth = i, random_state = 0, class_weight = 'balanced')
    model_tree.fit(x_train, y_train)
    train_score = model_tree.score(x_train, y_train)
    val_score = model_tree.score(x_val, y_val)
    return train_score, val_score
```

```python
def score(a):
    for i in range(1, 20):
        s1, s2 = learn_tree(a, t, i)
        print(i, s1, s2)
```
