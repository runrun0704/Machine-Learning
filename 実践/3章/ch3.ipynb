```python
import pandas as pd
```

```python
use_log = pd.read_csv('use_log.csv')
use_log
```

```python
customer_master = pd.read_csv('customer_master.csv')
customer_master
```

```python
class_master = pd.read_csv('class_master.csv')
class_master
```

```python
campaign_master = pd.read_csv('campaign_master.csv')
campaign_master
```

```python
tmp_df1 = customer_master.merge(use_log, how = 'left', on = 'customer_id')
tmp_df1
```

```python
tmp_df2 = tmp_df1.merge(campaign_master, how = 'left', on = 'campaign_id')
tmp_df2
```

```python
df = tmp_df2.merge(class_master, how = 'left', on = 'class')
df 
```

```python
df['is_deleted'].value_counts()
```
